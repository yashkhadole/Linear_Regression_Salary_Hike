```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
```

```python
df1 = pd.read_csv("/content/Salary_Data.csv")        # Salary dataset
df2 = pd.read_csv("/content/delivery_time.csv")      # Delivery dataset
```

```python
# Check Null Values
print("Null Values in Salary Dataset:\n", df1.isnull().sum())
```

```python
# Scatter plot
plt.figure()
sns.scatterplot(x=df1["YearsExperience"], y=df1["Salary"])
plt.title("Years Experience vs Salary")
plt.show()
```

```python
# Histogram
plt.figure()
sns.histplot(df1["Salary"], kde=True)
plt.title("Salary Distribution")
plt.show()
```

```python
# Boxplot
plt.figure()
sns.boxplot(data=df1)
plt.title("Salary Dataset Boxplot")
plt.show()
```

```python
# Heatmap
plt.figure()
sns.heatmap(df1.corr(), annot=True, cmap="coolwarm")
plt.title("Salary Dataset Correlation Heatmap")
plt.show()
```

```python
# 9. MODEL BUILDING - SALARY
print("Step 9: Building Linear Regression Model (Salary)...")

X = df1[["YearsExperience"]]
y = df1["Salary"]

model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)

r2 = r2_score(y, y_pred)
rmse = np.sqrt(mean_squared_error(y, y_pred))

print(f"Salary Model R2 Score: {round(r2,3)}")
print(f"Salary Model RMSE: {round(rmse,2)}\n")
```
